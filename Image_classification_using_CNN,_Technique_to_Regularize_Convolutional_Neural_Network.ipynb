# **`Data augmentation`**
Data augmentation is a technique used to artificially increase the size of a dataset by generating new data points based on existing data points. This is often done in machine learning to improve the model's performance and to reduce the risk of overfitting.

There are various methods for performing data augmentation, including adding noise to existing data points, rotating or shifting images, and applying various transformations to audio or text data.

Data augmentation can be particularly useful for tasks where the amount of available training data is limited, such as in image classification or natural language processing. By generating additional data points, the model can be exposed to a wider variety of inputs, which can improve its generalization ability.

However, it is important to ensure that the augmented data is representative of the underlying distribution of the data and is not introducing any biases. It is also important to carefully consider which augmentation techniques are appropriate for the specific task at hand.


# **`Batch normalization`**
Batch normalization is a technique that is commonly used in convolutional neural networks (CNNs) to improve the training process and increase the model's generalization ability. It works by normalizing the activations of a layer within a mini-batch of data, which can help to reduce the internal covariate shift and improve the optimization process.

Batch normalization has been shown to be effective at regularizing CNNs, which can help to reduce overfitting and improve the model's performance on unseen data. It can also help to stabilize the training process and make the model more robust to changes in the learning rate and other hyperparameters.

To implement batch normalization in a CNN, you can add a batch normalization layer after each convolutional or fully connected layer. The batch normalization layer takes in the activations of the previous layer and normalizes them using the mean and standard deviation calculated over the mini-batch. The normalized activations are then passed through an affine transformation (a linear transformation followed by a shift) to scale and shift the activations.

It is important to note that batch normalization should be used with care, as it can introduce additional hyperparameters and complexity to the model. It is also important to ensure that the batch normalization layers are properly trained and that the model is not over-regularized

#**`Learning rate scheduling`**
Learning rate scheduling is a technique used to adjust the learning rate of a neural network during training. It is often used to improve the model's convergence and to find a good balance between optimization and generalization.

In a convolutional neural network (CNN), the learning rate can be adjusted at different points in the training process, depending on the specific schedule being used. Some common types of learning rate schedules include:

* Constant learning rate: The learning rate remains the same throughout training.

* Step-based learning rate: The learning rate is decreased by a fixed amount at predetermined intervals.

* Exponential decay learning rate: The learning rate is decreased by a factor at each training step.

* 1/t decay learning rate: The learning rate is decreased by a factor of 1/t at each training step, where t is the current training step.

* Cyclical learning rate: The learning rate oscillates between two values over the course of training.

* Adaptive learning rate: The learning rate is adjusted based on the model's performance on the training data.

It is important to choose an appropriate learning rate schedule that is well-suited to the specific task and architecture of the CNN. It is also important to carefully tune the learning rate and other hyperparameters to achieve good performance.

#**`Weight Decay`**
Weight decay is a regularization technique used to prevent overfitting in neural networks, including convolutional neural networks (CNNs). It works by adding a penalty term to the loss function during training that encourages the model to have smaller weights.

In a CNN, weight decay is typically implemented as L2 regularization, which adds a penalty term to the loss function that is proportional to the sum of the squares of the weights. The penalty term is controlled by a hyperparameter called the weight decay coefficient, which determines the strength of the regularization.

Weight decay can help to reduce overfitting by constraining the model's ability to fit the training data too closely. It can also improve the generalization ability of the model by encouraging the weights to take on more reasonable values.

It is important to tune the weight decay coefficient carefully, as setting it too high can lead to underfitting, while setting it too low may not provide enough regularization. It is also important to consider the trade-off between regularization and model performance, as increasing the weight decay may improve generalization at the expense of a decrease in training accuracy.

#**`Gradient clipping`**
Gradient clipping is a technique used to prevent the gradients of a neural network from becoming too large or exploding during training. It is often used in conjunction with other optimization techniques, such as stochastic gradient descent (SGD), to improve the stability and convergence of the training process.

In a convolutional neural network (CNN), gradient clipping is typically applied to the gradients of the weights and biases. The gradients are clipped to a certain threshold, and any values that exceed the threshold are truncated to the threshold value.

Gradient clipping can help to stabilize the training process and prevent the gradients from becoming too large, which can cause the model to diverge or oscillate. It can also help to reduce the risk of overfitting, as large gradients can lead to rapid changes in the weights and biases, which can result in overfitting to the training data.

It is important to choose an appropriate threshold for gradient clipping, as setting it too high may not provide enough regularization, while setting it too low can slow down the training process. It is also important to consider the trade-off between stability and model performance, as increasing the gradient clipping threshold may improve stability at the expense of a decrease in training accuracy.



#**`Adaptive optimizer`**
This family of optimizers has been introduced to solve the issues of the gradient descent’s algorithms. Their most important feature is that they don’t require a tuning of the learning rate value. Actually some libraries — i.e. Keras — still let you the possibility to manually tune it for more advanced trials.

An adaptive optimizer is a type of optimization algorithm that adjusts its internal parameters during training based on the characteristics of the data and the model. These optimizers can often provide faster and more stable convergence compared to traditional optimization algorithms, such as stochastic gradient descent (SGD).

Some examples of popular adaptive optimizers include:

Adam (Adaptive Moment Estimation): Adam is a widely used optimizer that combines the benefits of both SGD and momentum optimization. It uses an exponentially weighted moving average of the gradients to adjust the learning rate on a per-parameter basis.

RMSprop (Root Mean Square Propagation): RMSprop is an optimizer that uses a moving average of the squared gradients to scale the learning rate. It is often used for training deep neural networks and can be more stable than SGD.

AdaDelta: AdaDelta is an optimizer that uses an exponentially weighted moving average of the squared gradients to scale the learning rate. It does not require the user to specify a learning rate, which can make it easier to use in practice.

Adagrad: Adagrad is an optimizer that scales the learning rate based on the sum of the squared gradients. It can be particularly effective for training large datasets and can adapt to the sparsity of the gradients.

Adaptive optimizers can be a useful tool for training neural networks, as they can often provide faster and more stable convergence compared to traditional optimization algorithms. However, it is important to carefully choose an appropriate optimizer for the specific task at hand and to tune the relevant hyperparameters to achieve good performance.

#**`Droput layer`**

Dropout is a regularization technique that is often used in convolutional neural networks (CNNs) to reduce the risk of overfitting. It works by randomly dropping out a subset of the units (neurons) in a layer during training, which can help to prevent the model from relying too heavily on any single unit.

To implement dropout in a CNN, you can add a dropout layer after each convolutional or fully connected layer. The dropout layer randomly sets a portion of the activations to zero during training, and the remaining activations are scaled by a factor of 1/(1-p), where p is the dropout rate. This ensures that the expected value of the activations remains the same, even when some units are dropped out.

Dropout can be a powerful tool for regularizing CNNs, as it can help to reduce overfitting by forcing the model to learn multiple redundant representations of the data. It can also improve the generalization ability of the model by encouraging the weights to take on more reasonable values.

It is important to choose an appropriate dropout rate and to tune it carefully, as setting it too high can lead to underfitting, while setting it too low may not provide enough regularization. It is also important to remember to turn off dropout during inference, as the activations are not scaled during this phase.

```python
def get_default_device():
    """Pick GPU if available, else CPU"""
    if torch.cuda.is_available():
        return torch.device('cuda')
    else:
        return torch.device('cpu')
    
def to_device(data, device):
    """Move tensor(s) to chosen device"""
    if isinstance(data, (list,tuple)):
        return [to_device(x, device) for x in data]
    return data.to(device, non_blocking=True)

class DeviceDataLoader():
    """Wrap a dataloader to move data to a device"""
    def __init__(self, dl, device):
        self.dl = dl
        self.device = device
        
    def __iter__(self):
        """Yield a batch of data after moving it to device"""
        for b in self.dl: 
            yield to_device(b, self.device)

    def __len__(self):
        """Number of batches"""
        return len(self.dl)
```

# 7: Configuring the model

## 7.1: Setting up Accuracy function and Image base class
Both are generic function and does not require any changes to be made regardless of the datasets. These are helper function to calculate accuracy and implements loss function to calculate the training and validation loss of the model.

```python
def accuracy(outputs, label):
  _,preds = torch.max(outputs, dim=1)
  return torch.tensor(torch.sum(preds==label).item()/len(preds))


class ImageClassificationBase(nn.Module):
    def training_step(self, batch):
        images, labels = batch 
        out = self(images)                  # Generate predictions
        loss = F.cross_entropy(out, labels) # Calculate loss
        return loss
    
    def validation_step(self, batch):
        images, labels = batch 
        out = self(images)                    # Generate predictions
        loss = F.cross_entropy(out, labels)   # Calculate loss
        acc = accuracy(out, labels)           # Calculate accuracy
        return {'val_loss': loss.detach(), 'val_acc': acc}
        
    def validation_epoch_end(self, outputs):
        batch_losses = [x['val_loss'] for x in outputs]
        epoch_loss = torch.stack(batch_losses).mean()   # Combine losses
        batch_accs = [x['val_acc'] for x in outputs]
        epoch_acc = torch.stack(batch_accs).mean()      # Combine accuracies
        return {'val_loss': epoch_loss.item(), 'val_acc': epoch_acc.item()}
    
    def epoch_end(self, epoch, result):
        print("Epoch [{}], train_loss: {:.4f}, val_loss: {:.4f}, val_acc: {:.4f}".format(epoch, result['train_loss'], result['val_loss'], result['val_acc']))
     

   
```

We use nn.Squential to chain together the layers of neural network. I have added comments in the code to simply the understanding. Note - Here I have implemented batch normalization using nn.BatchNorm2d module at the end of each layer.

```python

class Fruit360CnnModel(ImageClassificationBase):
    def __init__(self):
        super().__init__()
        self.network = nn.Sequential(
            nn.Conv2d(3, 16, kernel_size=2, padding=1), #input: 128* 3* 100*100
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(2, 2), #out: 128* 16* 50 *50

            nn.Conv2d(16, 32, kernel_size=2, stride=1, padding=1), 
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2, 2), #out  128* 32 X 25 X 25

            nn.Conv2d(32, 64, kernel_size=2, stride=1, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(5, 5), # 128* 64 * 5* 5

            nn.Flatten(), 
            nn.Dropout(0.3),
            nn.ReLU(),
            nn.Linear(64*5*5, 131))
        
    def forward(self, xb):
      return self.network(xb)
```

```python
model = Fruit360CnnModel()
model
```

## 7.3 Implementing Weight Decay, Gradient clipping, Adam optimizer while training


```python
@torch.no_grad()

def evaluate(model, val_loader):
    model.eval()
    outputs = [model.validation_step(batch) for batch in val_loader]
    return model.validation_epoch_end(outputs)

def get_lr(optimizer):
    for param_group in optimizer.param_groups:
        return param_group['lr']

def fit_one_cycle(epochs, max_lr, model, train_loader, val_loader, 
                  weight_decay=0, grad_clip=None, opt_func=torch.optim.SGD):
    torch.cuda.empty_cache()
    history = []
    
    # Set up cutom optimizer with weight decay
    optimizer = opt_func(model.parameters(), max_lr, weight_decay=weight_decay)
    # Set up one-cycle learning rate scheduler
    sched = torch.optim.lr_scheduler.OneCycleLR(optimizer, max_lr, epochs=epochs, 
                                                steps_per_epoch=len(train_loader))
    for epoch in range(epochs):
        # Training Phase 
        model.train()
        train_losses = []
        lrs = []
        for batch in train_loader:
            loss = model.training_step(batch)
            train_losses.append(loss)
            loss.backward()
            
            # Gradient clipping
            if grad_clip: 
                nn.utils.clip_grad_value_(model.parameters(), grad_clip)
            
            optimizer.step()
            optimizer.zero_grad()
            
            # Record & update learning rate
            lrs.append(get_lr(optimizer))
            sched.step()
        
        # Validation phase
        result = evaluate(model, val_loader)
        result['train_loss'] = torch.stack(train_losses).mean().item()
        result['lrs'] = lrs
        model.epoch_end(epoch, result)
        history.append(result)
    return history
```

## 8.3 Accuracy vs No of epochs


```python
def plot_accuracies(history):
    accuracies = [x['val_acc'] for x in history]
    plt.plot(accuracies, '-x')
    plt.xlabel('epoch')
    plt.ylabel('accuracy')
    plt.title('Accuracy vs. No. of epochs');

plot_accuracies(history)
```

## 8.4 Loss vs epochs


```python
def plot_losses(history):
    train_losses = [x.get('train_loss') for x in history]
    val_losses = [x['val_loss'] for x in history]
    plt.plot(train_losses, '-bx')
    plt.plot(val_losses, '-rx')
    plt.xlabel('epoch')
    plt.ylabel('loss')
    plt.legend(['Training', 'Validation'])
    plt.title('Loss vs. No. of epochs');

plot_losses(history)
```

## 8.5 Learning rate with batch no


```python
def plot_lrs(history):
    lrs = np.concatenate([x.get('lrs', []) for x in history])
    plt.plot(lrs)
    plt.xlabel('Batch no.')
    plt.ylabel('Learning rate')
    plt.title('Learning Rate vs. Batch no.');
```

# 11.Summary
Here's a summary of the different techniques used in this tutorial to improve our model performance and reduce the training time:

1. **Data augmentation:** We applied random transformations while loading images from the training dataset. Specifically, we will pad each image by 4 pixels, and then take a random crop of size 100 x 100 pixels, and then flip the image horizontally with a 50% probability.

2. **Batch normalization:** After each convolutional layer, we added a batch normalization layer, which normalizes the outputs of the previous layer. This is somewhat similar to data normalization, except it's applied to the outputs of a layer, and the mean and standard deviation are learned parameters.

3. **Learning rate scheduling:** Instead of using a fixed learning rate, we will use a learning rate scheduler, which will change the learning rate after every batch of training. There are many strategies for varying the learning rate during training, and we used the "One Cycle Learning Rate Policy".

4. **Weight Decay:** We added weight decay to the optimizer, yet another regularization technique which prevents the weights from becoming too large by adding an additional term to the loss function.

5. **Gradient clipping:** We also added gradient clippint, which helps limit the values of gradients to a small range to prevent undesirable changes in model parameters due to large gradient values during training.

6. **Adam optimizer:** Instead of SGD (stochastic gradient descent), we used the Adam optimizer which uses techniques like momentum and adaptive learning rates for faster training. There are many other optimizers to choose froma and experiment with.
